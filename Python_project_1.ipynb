```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#drop unrelated/blank column
df.drop(['Status','unnamed1'],axis = 1 , inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace= True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df[['Age','Orders','Amount']].describe()
```

Exploratory data analysis

Gender:

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

From above graphs we can see that most of buyers are females and even purchasing power of females are greater then mens.

Age:

```python
ax = sns.countplot(data = df  ,x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount v/s Age Groups
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

From above graphs we can see that most of buyers are of age group between 26-35 yrs females.

State:

```python
#total number of orders from top 10 states
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10
                                                                                                                        )
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

From above graphs we can see most of orders & total sales/amount are from Utter Pradesh, Maharastra and Karnataka respectively.

Marital Status:

```python
ax = sns.countplot(data = df  ,x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_state, hue = 'Gender')
```

From above graphs we can see that most of buyers are married women and they have high purchasing power.

Occupation:

```python
ax = sns.countplot(data = df  ,x = 'Occupation')

sns.set(rc={'figure.figsize':(18,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_occupation)
```

From above graphs we can see that most of buyers are from IT,Healthcare and Aviation sector.

Product category:

```python
ax = sns.countplot(data = df  ,x = 'Product_Category')

sns.set(rc={'figure.figsize':(18,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_product)
```

From above graphs we can see that most of sold products are from Food,Clothing and Electronic category.

Product Id:

```python
sales_productid = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_productid)
```

```python
#top 10 most sold product

gig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

Conclusion:

Married women age group 26-35 years from UP,Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food,Clothing and Electronic Category.
